count: 각 열에 있는 데이터의 총 개수
mean: 각 열의 평균값
std: 표준 편차로, 데이터가 평균값을 중심으로 얼마나 퍼져 있는지를 나타낸다
min: 각 열에서의 최솟값
25%: 1사분위수(25번째 백분위수)로, 전체 데이터의 하위 25% 지점의 값
50%: 중앙값(median)으로, 중간에 있는 값
75%: 3사분위수(75번째 백분위수)로, 전체 데이터의 상위 25% 지점의 값


```python
print(titanic.isnull().sum())
```

```python
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

print(titanic['age'].isnull().sum())
print(titanic['embarked'].isnull().sum())
```

```python
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic['alive'] = titanic['alive'].map({'no': 1, 'yes': 0})
titanic['embarked'] = titanic['embarked'].map({'C': 0, 'Q': 1, 'S': 2,})

print(titanic['sex'].head())
print(titanic['alive'].head())
print(titanic['embarked'].head())
```

```python
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

print(titanic['family_size'].head())
```
