### feature engineering

##### 3-1. 결측치 처리

```python
'''
  Age(나이)의 결측치는 중앙값으로, 
  Embarked(승선 항구)의 결측치는 최빈값으로 대체해주세요. 
  모두 대체한 후에, 대체 결과를 isnull() 함수와 sum()  함수를 이용해서 확인해주세요. 
'''
```

```python
# median()은 결측치의 중간값을 구하는 함수
# mode()는 최빈값을 구하는 함수
# 각각의 열에서 중간값과 최빈값을 구해 fillna()로 결측치를 채워줌
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])

print(titanic['age'].isnull().sum())
print(titanic['embarked'].isnull().sum())
```

###### 3-2. 수치형으로 인코딩

```python
'''
Sex(성별)를 남자는 0, 여자는 1로 변환해주세요. 
alive(생존여부)를 True는 1, False는 0으로 변환해주세요. 
Embarked(승선 항구)는 ‘C’는 0으로, Q는 1으로, ‘S’는 2로 변환해주세요. 
모두 변환한 후에, 변환 결과를 head 함수를 이용해 확인해주세요. 
```

```python
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic['alive'] = titanic['alive'].map({'no': 1, 'yes': 0})
titanic['embarked'] = titanic['embarked'].map({'C': 0, 'Q': 1, 'S': 2,})

print(titanic['sex'].head())
print(titanic['alive'].head())
print(titanic['embarked'].head())
```

##### 3-3. 새로운 feature 생성

```python
'''
SibSip(타이타닉호에 동승한 자매 및 배우자의 수), 
Parch(타이타닉호에 동승한 부모 및 자식의 수)를 통해서 family_size(가족크기)를 생성해주세요. 
새로운 feature를 head 함수를 이용해 확인해주세요. 
'''
```

```python
# sibsp (Siblings/Spouses Aboard) 타이타닉호에 동승한 자매 및 배우자의 수
# sibsp = 0: 형제자매나 배우자가 없는 경우
# sibsp = 1: 1명의 형제자매 또는 배우자가 함께 탑승한 경우

# parch (Parents/Children Aboard) 타이타닉호에 동승한 부모 및 자식의 수
# parch = 0: 부모나 자녀가 없는 경우
# parch = 2: 2명의 부모나 자녀가 함께 탑승한 경우

# 두 열 모두 승객 자신을 제외한 배우자, 형제자매, 부모, 자녀 등의 수를 나타내는 수이다.
# 그렇기에 가족의 수를 알기 위해서는 두 수에서 자신의 숫자를 더해줘야하기때문에
# SibSip + Parch +1 로 해줘야한다.
```

```python
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

print(titanic['family_size'].head())
```
