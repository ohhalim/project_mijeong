```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator, Vocab
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
import numpy as np
from tqdm import tqdm
import re
from collections import Counter

# 텍스트 전처리 함수
def preprocess_text(text):
    if isinstance(text, float):
        return ""
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    return text.strip()

# 수정된 vocab 생성 함수
def create_vocab(text_iterator):
    tokenizer = get_tokenizer('basic_english')
    counter = Counter()
    
    for text in tqdm(text_iterator, desc="Building vocabulary"):
        tokens = tokenizer(preprocess_text(text))
        counter.update(tokens)
    
    # 빈도수가 2 이상인 단어만 선택
    sorted_by_freq_tuples = sorted(counter.items(), key=lambda x: x[1], reverse=True)
    sorted_by_freq_tuples = [(word, freq) for word, freq in sorted_by_freq_tuples if freq >= 2]
    
    vocab_list = ['<unk>', '<pad>'] + [word for word, _ in sorted_by_freq_tuples]
    vocab_dict = {word: idx for idx, word in enumerate(vocab_list)}
    
    return vocab_dict

# 데이터셋 클래스
class ReviewDataset(Dataset):
    def __init__(self, reviews, ratings, vocab_dict, max_len=50):
        self.reviews = reviews
        self.ratings = ratings
        self.vocab_dict = vocab_dict
        self.max_len = max_len
        self.tokenizer = get_tokenizer('basic_english')

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, idx):
        # 텍스트 전처리 및 토큰화
        text = preprocess_text(self.reviews[idx])
        tokens = self.tokenizer(text)
        
        # 숫자로 변환 (없는 단어는 <unk> 토큰으로)
        ids = [self.vocab_dict.get(token, self.vocab_dict['<unk>']) for token in tokens]
        
        # 패딩
        if len(ids) > self.max_len:
            ids = ids[:self.max_len]
        else:
            ids = ids + [self.vocab_dict['<pad>']] * (self.max_len - len(ids))
        
        return (torch.tensor(ids, dtype=torch.long), 
                torch.tensor(self.ratings[idx] - 1, dtype=torch.long))

# LSTM 모델
class SimpleLSTM(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, output_dim):
        super(SimpleLSTM, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        
    def forward(self, text):
        embedded = self.embedding(text)
        lstm_out, (hidden, _) = self.lstm(embedded)
        return self.fc(hidden[-1])

def train_model(df, epochs=3):
    """
    전체 학습 프로세스를 관리하는 함수
    """
    # 1. 데이터 준비
    reviews = df['content'].values
    ratings = df['score'].values
    
    # 학습/테스트 분할
    train_reviews, test_reviews, train_ratings, test_ratings = train_test_split(
        reviews, ratings, test_size=0.2, random_state=42
    )
    
    # 2. 수정된 vocab 생성
    print("Creating vocabulary...")
    vocab_dict = create_vocab(train_reviews)
    print(f"Vocabulary size: {len(vocab_dict)}")
    
    # 3. 데이터셋 생성
    train_dataset = ReviewDataset(train_reviews, train_ratings, vocab_dict)
    test_dataset = ReviewDataset(test_reviews, test_ratings, vocab_dict)
    
    # 4. 데이터로더 설정
    train_dataloader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=16)
    
    # 5. 모델 초기화
    model = SimpleLSTM(
        vocab_size=len(vocab_dict),
        embed_dim=50,
        hidden_dim=64,
        output_dim=5
    )
    
    # 6. 학습 설정
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # 7. 학습 루프
    best_accuracy = 0
    
    for epoch in range(epochs):
        print(f'\nEpoch {epoch+1}/{epochs}')
        
        # 학습
        model.train()
        total_train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch in tqdm(train_dataloader, desc='Training'):
            reviews, ratings = batch
            
            optimizer.zero_grad()
            outputs = model(reviews)
            loss = criterion(outputs, ratings)
            
            loss.backward()
            optimizer.step()
            
            total_train_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            train_total += ratings.size(0)
            train_correct += (predicted == ratings).sum().item()
        
        train_accuracy = train_correct / train_total
        
        # 평가
        model.eval()
        total_test_loss = 0
        test_correct = 0
        test_total = 0
        
        with torch.no_grad():
            for batch in tqdm(test_dataloader, desc='Testing'):
                reviews, ratings = batch
                
                outputs = model(reviews)
                loss = criterion(outputs, ratings)
                
                total_test_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                test_total += ratings.size(0)
                test_correct += (predicted == ratings).sum().item()
        
        test_accuracy = test_correct / test_total
        
        print(f'Train Loss: {total_train_loss/len(train_dataloader):.3f}')
        print(f'Train Accuracy: {train_accuracy:.3f}')
        print(f'Test Loss: {total_test_loss/len(test_dataloader):.3f}')
        print(f'Test Accuracy: {test_accuracy:.3f}')
        
        # 최고 성능 모델 저장
        if test_accuracy > best_accuracy:
            best_accuracy = test_accuracy
            torch.save({
                'model_state_dict': model.state_dict(),
                'vocab_dict': vocab_dict,
            }, 'best_model.pt')
    
    return model, vocab_dict

def predict_review(model, vocab_dict, text):
    """
    새로운 리뷰에 대한 예측을 수행하는 함수
    """
    model.eval()
    tokenizer = get_tokenizer('basic_english')
    
    # 텍스트 전처리
    processed_text = preprocess_text(text)
    tokens = tokenizer(processed_text)
    
    # 토큰을 숫자로 변환
    ids = [vocab_dict.get(token, vocab_dict['<unk>']) for token in tokens]
    
    # 패딩
    if len(ids) > 50:
        ids = ids[:50]
    else:
        ids = ids + [vocab_dict['<pad>']] * (50 - len(ids))
    
    # 텐서 변환 및 예측
    with torch.no_grad():
        input_tensor = torch.tensor([ids], dtype=torch.long)
        output = model(input_tensor)
        predicted_score = torch.argmax(output).item() + 1
    
    return predicted_score

# 사용 예시
if __name__ == "__main__":
    print("모델 학습 시작...")
    model, vocab_dict = train_model(df, epochs=3)
    
    # 새로운 리뷰 예측
    test_review = "This is normal"
    predicted_score = predict_review(model, vocab_dict, test_review)
    print(f"\n리뷰: {test_review}")
    print(f"예측 평점: {predicted_score}")
```
